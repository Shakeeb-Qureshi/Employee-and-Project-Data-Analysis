### Task 2 :
#### The cost column in the dataframe “Project” has some missing values. Your task is to compute these missing values.
#### Replace the missing values by running average. You should use the “For” loop for this task.

```python
#Displaying Project DataFrame
PDF
```

```python
running_total = 0  # Initialize running total to 0 to store the sum of cost
count = 0  # Initialize count to 0 to keep track of the number of non-missing values
running_average = []  # Initialize an empty list to store the running average

# Calculate the running average
for cost in PDF[("Project DataFrame", "Cost")]:
    if not np.isnan(cost):  # Check if the cost value is not NaN
        running_total += cost  # Add the non-missing cost value to the running total
        count += 1  # Increment count for each non-missing value
    # Append the running average to the list, ensuring division by 0 is handled
    running_average.append(int(running_total / count) if count != 0 else 0)

# Replace missing values with the running average
for i in range(len(PDF)):
    if np.isnan(PDF.at[i, ("Project DataFrame", "Cost")]):  # Check if the cost value is NaN
        PDF.at[i, ("Project DataFrame", "Cost")] = running_average[i]  # Replace NaN with the running average
     
      
PDF
```

### Task 3 :
#### Split the name column in the Employee dataframe into two new columns “First Name”, and “Last Name” and remove the older “name” column.

```python
#Displaying Employee DataFrame
EDF
```

```python
# Extracting first name from the 'Name' column by splitting the string on whitespace and selecting the first element(index 0)
EDF[("Employee DataFrame", "First Name")] = EDF[("Employee DataFrame", "Name")].str.split(" ").str[0]

# Extracting last name from the 'Name' column by splitting the string on whitespace and selecting the second element(index 1)
EDF[("Employee DataFrame", "Last Name")] = EDF[("Employee DataFrame", "Name")].str.split(" ").str[1]

# Dropping the 'Name' column from the dataframe
EDF.drop(columns=[("Employee DataFrame", "Name")], inplace=True)

# Displaying the updated dataframe after dropping the 'Name' column
EDF
```

### Task 4 :
#### Join all three dataframes in one single dataframe. Name it “Final”

```python
# Join each level of the multi-level column index into a single string separated by a space
PDF.columns = [' '.join(col) for col in PDF.columns]

# Display the dataframe with the updated column labels
PDF
```

```python
# Join each level of the multi-level column index into a single string separated by a space
EDF.columns = [' '.join(col) for col in EDF.columns]

# Display the dataframe with the updated column labels
EDF
```

```python
# Join each level of the multi-level column index into a single string separated by a space
SDF.columns = [' '.join(col) for col in SDF.columns]

# Display the dataframe with the updated column labels
SDF
```

```python
# Merge PDF and EDF dataframes using an outer join on the common column 'Project DataFrame ID' and 'Employee DataFrame ID'
Final1 = pd.merge(PDF, EDF, how = 'outer', left_on = 'Project DataFrame ID', right_on = 'Employee DataFrame ID')

# Merge Final1 and SDF dataframes using an outer join on the common column 'Project DataFrame ID' and 'Seniority Level DataFrame ID'
Final = pd.merge(Final1, SDF, how='outer', left_on = 'Project DataFrame ID', right_on = 'Seniority Level DataFrame ID')

# Drop the columns 'Employee DataFrame ID' and 'Seniority Level DataFrame ID' from the merged dataframe Final
Final.drop(['Employee DataFrame ID', 'Seniority Level DataFrame ID'], axis=1, inplace=True)

# Display the final merged dataframe
Final
```

### Task 5 :
#### Add a new bonus column in the Final dataframe. Give a 5% bonus concerning project cost only to employees who have finished the projects.

```python
# Initialize the 'Bonus' column with default value 0 for all rows in the Final dataframe
Final["Bonus"] = 0

# Iterate over each row in the Final dataframe
for i, row in Final.iterrows():
    # Check if the project status is "Finished" for the current row
    if row[("Project DataFrame Status")] == "Finished":
        # Calculate bonus as 5% of the project cost and assign it to the 'Bonus' column for the current row
        Final.at[i, "Bonus"] = row[("Project DataFrame Cost")] * 0.05
    else:
        # If the project status is not "Finished", set the bonus value to 0 for the current row
        Final.at[i, "Bonus"] = 0

# Display the updated Final dataframe with bonus values
Final
```

### Task 6 :
#### Demote the designation level by 1, whose projects have status “fail”. Delete the employees record whose designation level is above 4.

```python
# Demote the designation level by 1 for employees with failed projects
Final.loc[Final[("Project DataFrame Status")] == "Failed", ("Seniority Level DataFrame Designation Level")] += 1

# Delete the records of employees with a designation level above 4
Final = Final[Final[("Seniority Level DataFrame Designation Level")] <= 4]

# Display the updated Final dataframe with bonus values
Final
```

### Task 7 :
#### Add “Mr.” and “Mrs.” to the first name column and drop the gender column.

```python
# Iterate over each row of the DataFrame
for i, row in Final.iterrows():
    # Check the gender
    if row[("Employee DataFrame Gender")] == "M":
        # If male, add "Mr." prefix to the first name
        Final.loc[i, ("Employee DataFrame First Name")] = "Mr. " + row[("Employee DataFrame First Name")]
    elif row[("Employee DataFrame Gender")] == "F":
        # If female, add "Mrs." prefix to the first name
        Final.loc[i, ("Employee DataFrame First Name")] = "Mrs. " + row[("Employee DataFrame First Name")]

# Drop the gender column
Final.drop(("Employee DataFrame Gender"), axis=1, inplace=True)


# Display the updated Final dataframe with bonus values
Final
```

### Task 8 :
#### Promote designation level by 1 for the employees whose age is more than 29 years using IF condition.

```python
# Iterate over each row in the DataFrame
for i, row in Final.iterrows():
    # Check if the age of the employee is greater than 29
    if row[("Employee DataFrame Age")] > 29:
        # If the condition is met, increase the designation level by 1
        Final.at[i, ("Seniority Level DataFrame Designation Level")] -= 1

# Display the updated Final dataframe with bonus values
Final
```

### Task 9 :
#### Add the cost of all projects for each Employee and save it in new dataframe “TotalProjCost” with three columns ID, First Name, and Total cost

```python
# Group by ID and First Name, then sum up the Cost for each group
TotalProjCost = Final.groupby([("Project DataFrame ID"), ("Employee DataFrame First Name")])[("Project DataFrame Cost")].sum().reset_index()

# Rename the columns for clarity
TotalProjCost.columns = ["ID", "First Name", "Total cost"]

# Print the new dataframe
TotalProjCost
```
